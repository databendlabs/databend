# Connecting to Databend

This notebook shows how to connect to Databend using the Python driver.

## Options
- **Local mode**: Run Databend locally with embedded storage
- **Cloud mode**: Connect to Databend Cloud (recommended for production)

## 1. Install the driver

```python
# Uncomment the line below if not installed
# !pip install "databend-driver[local]>=0.34.0"
```

## 2. Connect in local mode

Local mode runs an embedded Databend instance. No external server needed.

```python
from databend_driver import connect

# Connect to a local embedded instance
conn = connect("databend+local:///./local-state")

# Verify the connection
result = conn.query_row("SELECT 'Hello, Databend!' AS greeting")
print(result.values())
```

## 4. Useful connection info

```python
# Check Databend version
version = conn.query_row("SELECT version()")
print(f"Databend version: {version.values()}")

# List databases
databases = conn.query("SHOW DATABASES")
print(f"Databases: {databases.fetchall()}")
```
