# Basic SQL Operations

This notebook covers fundamental SQL operations in Databend:
- Creating tables
- Inserting data
- Querying with SELECT
- Filtering, sorting, and joining

```python
from databend_driver import connect

conn = connect("databend+local:///./local-state")
```

## 1. Create a table

```python
conn.execute("DROP TABLE IF EXISTS employees")
conn.execute("""
    CREATE TABLE employees (
        id INT,
        name VARCHAR,
        department VARCHAR,
        salary INT,
        hire_date DATE
    )
""")
print("Table created!")
```

## 2. Insert data

```python
conn.execute("""
    INSERT INTO employees VALUES
    (1, 'Alice', 'Engineering', 95000, '2022-01-15'),
    (2, 'Bob', 'Marketing', 72000, '2021-06-01'),
    (3, 'Charlie', 'Engineering', 105000, '2020-03-20'),
    (4, 'Diana', 'Sales', 68000, '2023-02-10'),
    (5, 'Eve', 'Marketing', 78000, '2022-08-05')
""")
print("Data inserted!")
```

## 3. Query data

```python
# Select all rows
result = conn.query("SELECT * FROM employees")
print(result.fetchall())
```

```python
# Filter with WHERE
result = conn.query("SELECT name, salary FROM employees WHERE salary > 75000")
print(result.fetchall())
```

```python
# Order by
result = conn.query("SELECT name, salary FROM employees ORDER BY salary DESC")
print(result.fetchall())
```

## 4. Aggregation

```python
result = conn.query("""
    SELECT department, 
           COUNT(*) AS count,
           AVG(salary) AS avg_salary
    FROM employees 
    GROUP BY department
    ORDER BY avg_salary DESC
""")
print(result.fetchall())
```

## 5. JOIN

```python
# Create a departments table
conn.execute("DROP TABLE IF EXISTS departments")
conn.execute("""
    CREATE TABLE departments (
        name VARCHAR,
        location VARCHAR
    )
""")
conn.execute("""
    INSERT INTO departments VALUES
    ('Engineering', 'San Francisco'),
    ('Marketing', 'New York'),
    ('Sales', 'Chicago')
""")

# JOIN query
result = conn.query("""
    SELECT e.name, e.department, d.location
    FROM employees e
    JOIN departments d ON e.department = d.name
""")
print(result.fetchall())
```

## 6. Convert to pandas DataFrame

```python
df = conn.query("SELECT * FROM employees").df()
print(type(df))
df.head()
```
